# 🐼 A. PANDAS

**Pandas** is the foundational Python library for data manipulation and analysis. Provides `DataFrame` and `Series` objects. Excels at reading, cleaning, and transforming tabular data (CSV, Excel, etc.).

| Function | Purpose |
|----------|--------|
| `pd.read_csv()` | Read CSV into DataFrame |
| `df.head(n)` | Top n rows (default 5) |
| `df.tail(n)` | Bottom n rows (default 5) |

```python
import pandas as pd

# --- Read CSV ---
data = pd.read_csv("https://media.geeksforgeeks.org/wp-content/uploads/nba.csv")

# --- head() : top 5 rows ---
print("HEAD:")
print(data.head())

# --- tail() : bottom 5 rows ---
print("\nTAIL:")
print(data.tail())

# --- sample() : random row ---
print("\nSAMPLE (2 rows):")
print(data.sample(n=2))

# --- info() : column names, dtypes, non-null count ---
print("\nINFO:")
data.info()

# --- describe() : statistical summary ---
print("\nDESCRIBE:")
print(data.describe())

# --- shape ---
print("\nShape:", data.shape)   # (rows, cols)

# --- column names ---
print("Columns:", list(data.columns))
```

# 🧹 DATA CLEANING (COMPULSORY)

**Data Cleaning** = process of detecting and correcting (or removing) corrupt, inaccurate, or missing records.

### Steps:
1. **Find missing values** → `isnull().sum()`
2. **Handle missing values** → `dropna()` or `fillna()`
3. **Remove duplicates** → `drop_duplicates()`
4. **Fix data types** → `astype()`
5. **Rename / Drop columns** → `rename()`, `drop()`
6. **Reset index** → `reset_index(drop=True)`

```python
import pandas as pd

data = pd.read_csv("https://media.geeksforgeeks.org/wp-content/uploads/nba.csv")

# 1. Check missing values
print("Missing values per column:")
print(data.isnull().sum())

# 2a. Drop rows with ANY missing value
data_dropped = data.dropna()
print("\nShape after dropna:", data_dropped.shape)

# 2b. Fill missing numeric values with column mean
data_filled = data.copy()
data_filled['Age'].fillna(data_filled['Age'].mean(), inplace=True)
data_filled['Salary'].fillna(data_filled['Salary'].median(), inplace=True)
data_filled['College'].fillna('Unknown', inplace=True)
print("\nMissing after fillna:")
print(data_filled.isnull().sum())

# 3. Remove duplicates
print("\nDuplicates:", data_filled.duplicated().sum())
data_clean = data_filled.drop_duplicates()

# 4. Check and fix data types
print("\nData types:")
print(data_clean.dtypes)

# 5. Rename a column
data_clean = data_clean.rename(columns={'Name': 'Player_Name'})

# 6. Reset index after cleaning
data_clean = data_clean.reset_index(drop=True)
print("\nCleaned data shape:", data_clean.shape)
print(data_clean.head())
```

# 📊 B. MATPLOTLIB — DATA VISUALIZATION (COMPULSORY)

**Matplotlib** is a multi-platform 2D plotting library built on NumPy. Created by John Hunter in 2002.

| Function | Plot Type |
|----------|----------|
| `plt.plot(x, y)` | Line plot |
| `plt.scatter(x, y)` | Scatter plot |
| `plt.bar(x, height)` | Bar chart |
| `plt.hist(data, bins)` | Histogram |
| `plt.pie(sizes)` | Pie chart |
| `plt.xlabel()` | Set X-axis label |
| `plt.ylabel()` | Set Y-axis label |
| `plt.title()` | Set plot title |
| `plt.legend()` | Show legend |
| `plt.show()` | Display plot |

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# --- 1. Line Plot ---
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y, label='Linear Function', color='blue')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.legend()
plt.show()

# --- 2. Scatter Plot ---
plt.scatter(x, y, c='blue', marker='o', label='Data Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.legend()
plt.show()

# --- 3. Bar Chart ---
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]
plt.bar(categories, values, width=0.6, align='center', color='green', label='Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.legend()
plt.show()

# --- 4. Histogram ---
data_hist = np.random.randn(1000)
plt.hist(data_hist, bins=30, color='purple', edgecolor='black', label='Normal Dist')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.legend()
plt.show()

# --- 5. Pie Chart ---
sizes = [25, 35, 20, 20]
labels = ['Python', 'Java', 'C++', 'R']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Language Popularity')
plt.show()
```

```python
# --- Visualization on NBA dataset (Data Cleaning + Visualization combined) ---
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("https://media.geeksforgeeks.org/wp-content/uploads/nba.csv")
data.dropna(inplace=True)   # Clean first!

# Top 5 teams by player count
team_counts = data['Team'].value_counts().head(5)
plt.bar(team_counts.index, team_counts.values, color='steelblue')
plt.xlabel('Team')
plt.ylabel('Number of Players')
plt.title('Top 5 NBA Teams by Player Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Salary distribution
plt.hist(data['Salary'], bins=20, color='orange', edgecolor='black')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('NBA Player Salary Distribution')
plt.show()

# Age vs Salary scatter
plt.scatter(data['Age'], data['Salary'], alpha=0.5, color='teal')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age vs Salary')
plt.show()
```

# 🔬 C. SCIPY

**SciPy (Scientific Python)** — open-source scientific computing library built on NumPy. Most used scientific library after GNU Scientific Library.

| Module | Purpose |
|--------|---------|
| `scipy.integrate` | Numerical integration (quad, ODE) |
| `scipy.linalg` | Linear algebra (solve, SVD, eigenvalues) |
| `scipy.special` | Special functions (factorial, Bessel, gamma) |
| `scipy.stats` | Statistics (distributions, tests, rank) |
| `scipy.signal` | Signal processing (filter, FFT, correlate) |

```python
import numpy as np

# --- i. Integrate: numerical integration using quad ---
from scipy.integrate import quad

def f(x):
    return np.exp(-x)

result, error = quad(f, 0, 1)   # integrate f(x) from 0 to 1
print(f"Integral of e^(-x) from 0 to 1: {result:.4f}, error: {error:.2e}")
# Expected: ~0.6321

# --- ii. Linear Algebra: solve system of equations ---
from scipy.linalg import solve

# Solve: 3x + 2y = 1
#         x + 4y = 2
A = np.array([[3, 2], [1, 4]])
B = np.array([[1], [2]])
x = solve(A, B)
print("\nSolution for linear system Ax=B:")
print(x)

# --- iii. Special: factorial ---
from scipy.special import factorial

n = 5
result = factorial(n)
print(f"\nFactorial of {n} = {result}")

# --- iv. Stats: rank data ---
from scipy.stats import rankdata

my_array = np.array([3, 5, 2, 1, 9, 9])
ranks = rankdata(my_array)
print("\nRanks:", ranks)
# Tied values (9,9) get average rank 5.5

# --- v. Signal: cross-correlation ---
from scipy.signal import correlate

seq1 = np.array([1, 2, 3, 4, 5, 6])
seq2 = np.array([4, 3, 2, 1])
correlation = correlate(seq1, seq2)
print("\nCross-correlation:", correlation)
```

# 🤖 D. SCIKIT-LEARN (Sklearn)

Started as Google Summer of Code project by David Cournapeau. Built on NumPy, SciPy, Matplotlib. Core algorithms in Cython for performance. SVM/Logistic regression use LIBSVM/LIBLINEAR wrappers.

| Feature | Function |
|---------|----------|
| Built-in Datasets | `datasets.load_iris()`, `load_diabetes()` |
| Train/Test Split | `train_test_split(X, y, test_size=0.2)` |
| Linear Regression | `LinearRegression().fit(X_train, y_train)` |
| Logistic Regression | `LogisticRegression().fit(X_train, y_train)` |
| Decision Tree | `DecisionTreeClassifier().fit(X_train, y_train)` |
| Metrics | `confusion_matrix`, `classification_report`, `r2_score`, `mean_squared_error` |

```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import (confusion_matrix, classification_report,
                              mean_squared_error, r2_score)

# --- i. Load built-in dataset ---
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
print("Iris dataset head:")
print(df.head())

# --- ii. Train/Test Split ---
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(f"\nTrain: {X_train.shape}, Test: {X_test.shape}")

# --- iii. Linear Regression (on Boston/diabetes dataset) ---
diabetes = datasets.load_diabetes()
Xd, yd = diabetes.data, diabetes.target
Xd_train, Xd_test, yd_train, yd_test = train_test_split(Xd, yd, test_size=0.2, random_state=0)

lr = LinearRegression()
lr.fit(Xd_train, yd_train)
yd_pred = lr.predict(Xd_test)
print(f"\nLinear Regression R2 Score: {r2_score(yd_test, yd_pred):.4f}")
print(f"MSE: {mean_squared_error(yd_test, yd_pred):.4f}")

# --- iv. Logistic Regression (on Iris) ---
log_reg = LogisticRegression(max_iter=200)
log_reg.fit(X_train, y_train)
y_pred_lr = log_reg.predict(X_test)
print("\nLogistic Regression Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))

# --- v. Decision Tree ---
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
print("Decision Tree Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_dt))
print(f"Decision Tree Accuracy: {(y_pred_dt == y_test).mean()*100:.2f}%")
```

# 🔢 E. NUMPY

**NumPy (Numerical Python)** — foundation for scientific computing in Python. Processes n-dimensional arrays of same data type. Vectorized operations = faster execution.

**Syntax:** `numpy.function_name(array, parameters)`

| Category | Functions |
|----------|-----------|
| Creation | `np.array()`, `np.zeros()`, `np.ones()`, `np.arange()`, `np.linspace()` |
| Manipulation | `np.reshape()`, `np.transpose()`, `np.flatten()`, `np.concatenate()` |
| Math | `np.add()`, `np.subtract()`, `np.multiply()`, `np.dot()`, `np.sqrt()` |
| Statistics | `np.mean()`, `np.median()`, `np.std()`, `np.var()`, `np.min()`, `np.max()` |

```python
import numpy as np

# --- i. Array Creation ---
arr1 = np.array([1, 3, 5])
print("np.array():", arr1)

print("np.zeros(3):", np.zeros(3))
print("np.ones((2,3)):\n", np.ones((2,3)))
print("np.arange(0,10,2):", np.arange(0, 10, 2))
print("np.linspace(0,1,5):", np.linspace(0, 1, 5))

# --- ii. Array Manipulation ---
array1 = np.array([1, 3, 5, 7, 9, 11])
array2 = np.reshape(array1, (2, 3))
print("\nReshaped array:\n", array2)
print("Transposed:\n", array2.T)
print("Flattened:", array2.flatten())

# --- iii. Mathematical Functions ---
a = np.array([1, 2, 3, 4, 5])
b = np.array([4, 9, 16, 25, 36])
print("\nnp.add:", np.add(a, b))
print("np.sqrt:", np.sqrt(b))
print("np.dot (dot product):", np.dot(a, a))  # 1+4+9+16+25=55

# --- iv. Statistical Functions ---
marks = np.array([76, 78, 81, 66, 85])
print("\nMarks:", marks)
print("Mean:", np.mean(marks))       # 77.2
print("Median:", np.median(marks))   # 78.0
print("Std Dev:", np.std(marks))
print("Variance:", np.var(marks))
print("Min:", np.min(marks), "Max:", np.max(marks))
```
